##### Some DataFrames, such as `df_protein_calories`, have the 'Year' column presented as a period rather than a single year. To address this, I created a function to transform these periods into a single year for consistency in the analysis.

```python
# subject: If provided, only rows with this subject will be updated.
def get_middle_year(variable, subject = None):
    def calculating(period):

        # Takes a period string (e.g., '2000-2005'), splits it into start and end years, converts them to integers, and calculates the midpoint.
        start_year, end_year = map(int, period.split('-'))
        result = (start_year + end_year) // 2
        return result
    
    # If subject is provided, only rows where the Subject column matches the given subject are updated.
    if subject is None:
        variable['Year'] = variable['Year'].apply(calculating)
    else:
        variable.loc[variable['Subject'] == subject, 'Year'] = variable.loc[variable['Subject'] == subject, 'Year'].apply(calculating)
    return variable

df_protein_calories = get_middle_year(df_protein_calories, 'avg protein supply')
df_food_situation = get_middle_year(df_food_situation, 'N. of obese adults')
df_food_situation = get_middle_year(df_food_situation, 'N. of people undernourished')
df_food_insecurity = get_middle_year(df_food_insecurity)
```

##### I am creating a function to categorize all data, including sex, year, and the respective countries based on their rankings. I am also handling food insecurity separately, as it is the only category divided by sex and contains data only from 2015 to 2022.

```python
def m_or_f(variable, list, sex, year):
    df = variable.copy()
    words_to_exclude_m = ['female', 'total', 'people']

    # Create a boolean mask for rows that should be included/excluded based on gender
    if sex == 'male':
        # The mask excludes rows where the 'Subject' column contains any of the words in words_to_exclude_m (e.g., 'female', 'total', 'people')
        mask = ~df['Subject'].str.contains('|'.join(words_to_exclude_m), case=False, na=False)
    elif sex == 'female':
        # For females: The mask includes rows where the 'Subject' column contains the word 'female'
        mask = df['Subject'].str.contains('female', case=False, na=False)
    else:
        print('error')
        return df 

    # Apply the mask to filter the DataFrame
    df = df[mask]

    # Filtering the DataFrame based on year and list of countries
    df = df[
        (df['Year'] == year) & 
        (df['Country'].isin(list))
    ]

    # Ensure the 'Country' column follows the specified order
    df['Country'] = pd.Categorical(df['Country'], categories=list, ordered=True)

    # Sort the DataFrame by the 'Country' column
    df = df.sort_values('Country')
    df = df.reset_index(drop=True)  
    return df
```

```python
# Function to filter and sort data based on country list and year
def rest(variable, country_list, year):
    df = variable.copy()  
    df = df[
        (df['Year'] == year) &  # Filter for the given year
        (df['Country'].isin(country_list))  # Filter for countries in the list
    ]

    # Ensure the 'Country' column follows the specified order
    df['Country'] = pd.Categorical(df['Country'], categories=country_list, ordered=True)

    # Sort the DataFrame by the 'Country' column and reset the index
    df = df.sort_values('Country').reset_index(drop=True)
    return df

# Define categories, positions, and years
categories = ['obesity', 'protein_calories', 'food_situation']
positions = ['top_b_b', 'top_b_60', 'bottom_b_b', 'bottom_b_60']
years = [2000, 2010, 2015, 2021]

# Initialize a nested dictionary to hold DataFrames for each category, position, and year
data_dfs = {cat: {pos: {} for pos in positions} for cat in categories}

# Loop over years, categories, and positions to populate the dictionary with filtered DataFrames
for year in years:
    for category in categories:
        for position in positions:
            list_var_name = f"list_{position}"  # Variable name for the country list
            df_var_name = f"df_{category}"  # Variable name for the DataFrame

            # Access the country list and DataFrame using globals() and handle missing values
            list_var = globals().get(list_var_name, {}).get(year, [])
            df_var = globals().get(df_var_name)

            # If the DataFrame is found, apply the 'rest' function and store the result
            if df_var is not None:
                data_dfs[category][position][year] = rest(df_var, list_var, year)

# Function to concatenate DataFrames for different years
def concat_dfs(data_dict):
    return pd.concat(data_dict.values(), ignore_index=True)

# Create a dictionary that holds the concatenated DataFrames for each category and position
all_data = {
    'obesity': {
        'top_b_b': concat_dfs(data_dfs['obesity']['top_b_b']),
        'top_b_60': concat_dfs(data_dfs['obesity']['top_b_60']),
        'bottom_b_b': concat_dfs(data_dfs['obesity']['bottom_b_b']),
        'bottom_b_60': concat_dfs(data_dfs['obesity']['bottom_b_60']),
    },
    'protein_calories': {
        'top_b_b': concat_dfs(data_dfs['protein_calories']['top_b_b']),
        'top_b_60': concat_dfs(data_dfs['protein_calories']['top_b_60']),
        'bottom_b_b': concat_dfs(data_dfs['protein_calories']['bottom_b_b']),
        'bottom_b_60': concat_dfs(data_dfs['protein_calories']['bottom_b_60']),
    },
    'food_situation': {
        'top_b_b': concat_dfs(data_dfs['food_situation']['top_b_b']),
        'top_b_60': concat_dfs(data_dfs['food_situation']['top_b_60']),
        'bottom_b_b': concat_dfs(data_dfs['food_situation']['bottom_b_b']),
        'bottom_b_60': concat_dfs(data_dfs['food_situation']['bottom_b_60']),
    }
}
```

```python
def ranking_countries(raw, dataset):
    
    variable = dataset.copy()

    # Create a mapping of (Country, Year) pairs to their corresponding ranking values from the 'raw' DataFrame
    ranking_map = raw.set_index(['Country', 'Year'])['Ranking'].to_dict()

    # Set the index of 'variable' DataFrame to (Country, Year)
    variable = variable.set_index(['Country', 'Year'])

    # Map the rankings from 'ranking_map' to the 'variable' DataFrame using the (Country, Year) pairs
    variable['Ranking'] = variable.index.map(ranking_map)

    variable.reset_index(inplace=True)

    return variable
```
