```python
# median increase per year
def median_per_year(variable):
    median = variable.groupby(['Year'])['Year_Expected'].median().reset_index()
    rows = median['Year_Expected'].tolist()

    diff_list = []
    for i in range(0, len(rows) - 1): 
        diff = rows[i] - rows[i+1]  
        diff_list.append(diff)

    diff_list = [abs(value) for value in diff_list]
    diff_list = pd.DataFrame(diff_list, columns=['difference'])
    diff_list = diff_list['difference'].median()
    return diff_list

btb = median_per_year(top_b_at_birth)
bbb = median_per_year(bottom_b_at_birth)
bt60 = median_per_year(top_b_at_60)
bb60 = median_per_year(bottom_b_at_60)

mtb = median_per_year(top_m_at_birth)
mbb = median_per_year(bottom_m_at_birth)
ftb = median_per_year(top_f_at_birth)
fbb = median_per_year(bottom_f_at_birth)

mt60 = median_per_year(top_m_at_60)
mb60 = median_per_year(bottom_m_at_60)
ft60 = median_per_year(top_f_at_60)
fb60 = median_per_year(bottom_f_at_60)
```

```python
import statistics
def median_life_expec(variable):
    temp_df = variable.groupby(['Country', 'categorize', 'Country_Code'])['Year_Expected'].median().reset_index()
    temp_df = temp_df.sort_values(by= 'Year_Expected', ascending=False).head(10)
    country_list = temp_df['Country'].tolist()
    year_expected_list = temp_df['Year_Expected'].tolist()
    temp_df = statistics.median(year_expected_list)
    return country_list, temp_df

tbabl, tbab = median_life_expec(top_b_at_birth)
bbabl, bbab = median_life_expec(bottom_b_at_birth)
```

```python
def max(variable, category, filtering):
    
    category_map = {
        'c': 'Communicable, maternal, perinatal and nutritional conditions',
        'n': 'Noncommunicable diseases',
        'i': 'Injuries'
    }
    
    if category in category_map:
        full_category_name = category_map[category]
        check = variable[variable['Category'] == full_category_name]
        

    check = check[check['Country'].isin(filtering)]
    check = check.groupby(['Category', 'Sub_Category', 'Death_Cause', 'categorize'])[['Value']].median().reset_index()
    check['Value']= check['Value']*1000
    temp_df = check.groupby('Category')['Value'].sum().reset_index()
    total_value_injuries = temp_df.loc[temp_df['Category'] == full_category_name, 'Value'].values[0]
    check['Percentage'] = (check['Value'] / total_value_injuries) * 100
    check = check.loc[check.groupby(['Sub_Category'])['Value'].idxmax()].reset_index(drop=True)
    return check 
```
