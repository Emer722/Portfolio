```python
def process_death(file_path, sheet_name):
    df_death = pd.read_excel(file_path, sheet_name=sheet_name)
    df_cleaned = df_death.copy()

    # Clean the data
    rows_to_drop = [0, 1, 2, 3, 4]
    df_cleaned = df_cleaned.drop(rows_to_drop)
    df_cleaned = df_cleaned.dropna(how='all')

    # Set new column names
    new_column_names = df_cleaned.iloc[0]
    df_cleaned.columns = new_column_names
    df_cleaned = df_cleaned[1:]

    # Rename specific columns
    new_columns = df_cleaned.columns.tolist()
    new_columns[2] = 'e'
    new_columns[4] = 'r'
    new_columns[5] = 't'
    df_cleaned.columns = new_columns
    
    return df_cleaned
```

```python
def Death_cause(variable, file_path, sheet_name, start, end, rows, Death_Cause, columns, numbers, rows_to_drop_male, diseases, check_if_male):
    
    # Process the death data based on whether 'Males' or not
    if check_if_male == 'Yes':
        
        variable = process_death(file_path, sheet_name).copy()
        # Filter the data for males only
        variable = variable[variable['Sex'] == 'Males']
        # Reset index for the DataFrame
        variable = variable.reset_index()  
    elif check_if_male == 'No':
        
        variable = process_death(file_path, sheet_name).copy()
    else:
        
        print('error')
        
    # Slice the DataFrame based on start and end indices
    variable = variable.iloc[start:end]
    # Drop rows where 'GHE cause' is in the 'rows' list
    variable = variable[~variable['GHE cause'].isin(rows)]
    # Drop rows where 't' is in the 'Death_Cause' list
    variable = variable[~variable['t'].isin(Death_Cause)]
    # Drop specified columns by their names
    variable = variable.drop(numbers)
    # Drop specified columns by their names
    variable = variable.drop(columns=columns)

    # If the data is for males, drop additional rows based on 'rows_to_drop_male'
    if check_if_male == 'Yes':
        variable = variable[~variable['index'].isin(rows_to_drop_male)]
    elif check_if_male == 'No':
        pass  
    else:
        
        print('error')

    # Remove columns with NaN values
    variable = variable.loc[:, variable.columns.notna()]
    # Reset the index of the DataFrame
    variable = variable.reset_index()
    # Add 'Category', 'Sub_Category', 'Death_Cause', 'Type' columns from 'diseases' DataFrame
    variable[['Category', 'Sub_Category', 'Death_Cause', 'Type']] = diseases[['Category', 'Sub_Category', 'Death_Cause', 'Type']]
    # Reshape the DataFrame from wide to long format
    variable = variable.melt(id_vars=['Sex', 'Category', 'Sub_Category', 'Death_Cause', 'Type'], value_vars=list_of_country_names, var_name='Country', value_name='Value')
    # Merge with 'countries' DataFrame to add country codes and country name
    variable = pd.merge(variable, countries, on='Country', how='left')

    # Reorder columns and replace 'Persons' with 'Both' in the 'Sex' column
    variable = variable[['Sex', 'Country', 'Country_code', 'Category', 'Sub_Category', 'Death_Cause', 'Type', 'Value']]
    variable['Sex'] = variable['Sex'].replace({'Persons': 'Both'})
    
    
    return variable
```

```python
def process_file(file_path, sheet_name):
    combined_df = pd.DataFrame()  # Initialize an empty DataFrame to hold the combined results

    # Extract the year from the file name
    year = file_path.split('_')[-1].split('.')[0]
    
    # Process data for both genders across different categories
    # Communicable diseases for both genders
    df_Communicable_both_gender = Death_cause(
        pd.DataFrame(),  # Pass an empty DataFrame as the initial variable
        file_path,       # Path to the file
        sheet_name,      # Name of the sheet to read from
        0, 66,           # Row indices to select (start: 0, end: 66)
        ['A.', 'B.', 'C.', 'D.', 'E.'],  # Rows to exclude based on 'GHE cause'
        ['STDs excluding HIV', 'Childhood-cluster diseases', 'Hepatitis', 'Parasitic and vector diseases', 'Intestinal nematode infections'],  # Diseases to exclude
        ['GHE code', 'GHE cause', 'Country or area\n(See Notes for explanation of colour codes)'],  # Columns to drop
        [6, 7, 8, 9, 10],  # Columns to drop by index
        [],  # No additional rows to drop for this category
        df_Communicable_diseases,  # Diseases DataFrame with detailed classifications
        'No'  # Process data for both genders (not filtered by sex)
    )

    # Non-communicable diseases for both genders
    df_Non_Communicable_both_gender = Death_cause(
        pd.DataFrame(),  # Pass an empty DataFrame as the initial variable
        file_path,       # Path to the file
        sheet_name,      # Name of the sheet to read from
        66, 203,         # Row indices to select (start: 66, end: 203)
        ['A.', 'B.', 'C.', 'D.', 'E.', 'F.', 'G.', 'H.', 'I.', 'J.', 'K.', 'L.', 'M.', 'N.', 'O.', 'P.'],  # Rows to exclude based on 'GHE cause'
        ['Mouth and oropharynx cancers', 'Liver cancer', 'Melanoma and other skin cancers', 'Lymphomas, multiple myeloma', 'Drug use disorders', 'Childhood behavioural disorders', 'Stroke', 'Cirrhosis of the liver', 'Kidney diseases'],  # Diseases to exclude
        ['GHE code', 'GHE cause', 'Country or area\n(See Notes for explanation of colour codes)', 'e', 'r', 't'],  # Columns to drop
        [72],  # Columns to drop by index
        [],  # No additional rows to drop for this category
        df_Non_Communicable_diseases,  # Diseases DataFrame with detailed classifications
        'No'  # Process data for both genders (not filtered by sex)
    )

    # Injuries for both genders
    df_Injury_both_gender = Death_cause(
        pd.DataFrame(),  # Pass an empty DataFrame as the initial variable
        file_path,       # Path to the file
        sheet_name,      # Name of the sheet to read from
        203, 218,        # Row indices to select (start: 203, end: 218)
        ['A.', 'B.', 'C.'],  # Rows to exclude based on 'GHE cause'
        [],  # No diseases to exclude
        ['GHE code', 'GHE cause', 'Country or area\n(See Notes for explanation of colour codes)', 'e', 'r', 't'],  # Columns to drop
        [209],  # Columns to drop by index
        [],  # No additional rows to drop for this category
        df_Injury_diseases,  # Diseases DataFrame with detailed classifications
        'No'  # Process data for both genders (not filtered by sex)
    )

    # Combine the DataFrames for all categories and both genders
    Death_cause_both_gender = pd.concat([df_Communicable_both_gender, df_Non_Communicable_both_gender, df_Injury_both_gender])

    # Process data for males only
    df_Communicable_male = Death_cause(
        pd.DataFrame(),  # Pass an empty DataFrame as the initial variable
        file_path,       # Path to the file
        sheet_name,      # Name of the sheet to read from
        2, 64,           # Row indices to select (start: 2, end: 64)
        ['A.', 'B.', 'C.', 'D.', 'E.'],  # Rows to exclude based on 'GHE cause'
        ['STDs excluding HIV', 'Childhood-cluster diseases', 'Hepatitis', 'Parasitic and vector diseases', 'Intestinal nematode infections'],  # Diseases to exclude
        ['GHE code', 'GHE cause', 'Country or area\n(See Notes for explanation of colour codes)'],  # Columns to drop
        [],  # No columns to drop by index
        [228, 231, 240, 247, 252, 266],  # Rows to drop for males
        df_Communicable_diseases,  # Diseases DataFrame with detailed classifications
        'Yes'  # Process data only for males
    )

    df_Non_Communicable_male = Death_cause(
        pd.DataFrame(),  # Pass an empty DataFrame as the initial variable
        file_path,       # Path to the file
        sheet_name,      # Name of the sheet to read from
        65, 200,         # Row indices to select (start: 65, end: 200)
        ['A.', 'B.', 'C.', 'D.', 'E.', 'F.', 'G.', 'H.', 'I.', 'J.', 'K.', 'L.', 'M.', 'N.', 'O.', 'P.'],  # Rows to exclude based on 'GHE cause'
        ['Mouth and oropharynx cancers', 'Liver cancer', 'Melanoma and other skin cancers', 'Lymphomas, multiple myeloma', 'Drug use disorders', 'Childhood behavioural disorders', 'Stroke', 'Cirrhosis of the liver', 'Kidney diseases'],  # Diseases to exclude
        ['GHE code', 'GHE cause', 'Country or area\n(See Notes for explanation of colour codes)', 'e', 'r', 't'],  # Columns to drop
        [],  # No columns to drop by index
        [],  # No additional rows to drop for this category
        df_Non_Communicable_diseases,  # Diseases DataFrame with detailed classifications
        'Yes'  # Process data only for males
    )

    df_Injury_male = Death_cause(
        pd.DataFrame(),  # Pass an empty DataFrame as the initial variable
        file_path,       # Path to the file
        sheet_name,      # Name of the sheet to read from
        202, 215,        # Row indices to select (start: 202, end: 215)
        ['A.', 'B.', 'C.'],  # Rows to exclude based on 'GHE cause'
        [],  # No diseases to exclude
        ['GHE code', 'GHE cause', 'Country or area\n(See Notes for explanation of colour codes)', 'e', 'r', 't'],  # Columns to drop
        [],  # No columns to drop by index
        [],  # No additional rows to drop for this category
        df_Injury_diseases,  # Diseases DataFrame with detailed classifications
        'Yes'  # Process data only for males
    )

    # Combine the DataFrames for all categories and males
    Death_cause_male = pd.concat([df_Communicable_male, df_Non_Communicable_male, df_Injury_male])

    # Create a female DataFrame based on both data
    Death_cause_female = Death_cause_male.copy()
    Death_cause_female['Sex'] = 'Females'  # Update the 'Sex' column to 'Females'
    Death_cause_female['Value'] = Death_cause_both_gender['Value'] - Death_cause_male['Value']  # Calculate the female values
    Death_cause_female['Value'] = Death_cause_female['Value'].abs()  # Ensure values are positive

    # Combine DataFrames for both genders and both sexes
    combined_df = pd.concat([Death_cause_both_gender, Death_cause_male, Death_cause_female], ignore_index=True)
    combined_df['Year'] = year  # Add the year to the final DataFrame

   
    return combined_df
```

```python
def combine_data(file_paths, ages):
    combined_df = pd.DataFrame()  

    # Iterate through each file path
    for file_path in file_paths:
        # Iterate through each age group
        for age in ages:
            sheet_name = str(age)  # Convert the age group to string for the sheet name
            
            # Process the file and sheet for the given age group
            df_age = process_file(file_path, sheet_name)
            
            # Add a new column 'Age_Group' to the DataFrame to specify the age group
            df_age['Age_Group'] = age
            
            # Concatenate the current DataFrame with the combined DataFrame
            combined_df = pd.concat([combined_df, df_age], ignore_index=True)
        
    
    return combined_df


# List of file paths to process
file_paths = [
    "C://Users/Emerson Wu Li/Desktop/longevity/ghe2021_deaths_bycountry_2021.xlsx",
    "C://Users/Emerson Wu Li/Desktop/longevity/ghe2021_deaths_bycountry_2015.xlsx",
    "C://Users/Emerson Wu Li/Desktop/longevity/ghe2021_deaths_bycountry_2010.xlsx",
    "C://Users/Emerson Wu Li/Desktop/longevity/ghe2021_deaths_bycountry_2000.xlsx"
]

# List of age groups to process
ages = ['0-4', '5-14', '15-29', '30-49', '50-59', '60-69', '70+']

# Combine data from all files and age groups
df_combined = combine_data(file_paths, ages)

# Ensure the 'Year' column is of integer type
df_combined['Year'] = df_combined['Year'].apply(int)
```

```python
def rest(variable, country_list, year):
    df = variable.copy()
    
    # Filter the DataFrame for the specified year and countries in the country_list
    df = df[
        (df['Year'] == year) & 
        (df['Country'].isin(country_list))
    ]
    
    # Ensure the 'Country' column follows the specified order
    df['Country'] = pd.Categorical(df['Country'], categories=country_list, ordered=True)
    
    # Sort the DataFrame by the 'Country' column and reset the index
    df = df.sort_values('Country').reset_index(drop=True)
    
    return df

# Define the categories and positions for filtering data
categories = ['birth_both', 'birth_male', 'sixty_both', 'sixty_male', 'birth_female', 'sixty_female', 'rests_male', 'rests_female', 'rests_both']
positions = ['top_m_b', 'top_m_60', 'bottom_m_b', 'bottom_m_60', 'top_f_b', 'top_f_60', 'bottom_f_b', 'bottom_f_60', 'top_b_b', 'top_b_60', 'bottom_b_b', 'bottom_b_60']
years = [2000, 2010, 2015, 2021]

# Initialize a nested dictionary to store DataFrames
data_dfs = {cat: {pos: {} for pos in positions} for cat in categories}

# Populate the dictionary with DataFrames filtered by year and category
for year in years:
    for category in categories:
        for position in positions:
            list_var_name = f"list_{position}"  # Construct the variable name for country list
            df_var_name = category  # Use the category name to get the DataFrame
            
            # Access the variables using globals() with error handling
            list_var = globals().get(list_var_name, {}).get(year, [])  # Get the list of countries for the given year
            df_var = globals().get(df_var_name)  # Get the DataFrame for the given category
            
            # Applying the 'rest' function to filter data
            data_dfs[category][position][year] = rest(df_var, list_var, year)

def concat_dfs(data_dict):
    # Concatenate all DataFrames in the dictionary into a single DataFrame
    return pd.concat(data_dict.values(), ignore_index=True)

# Combine all filtered DataFrames into a single dictionary
all_data = {
    'birth_male': {
        'top_m_b': concat_dfs(data_dfs['birth_male']['top_m_b']),
        'bottom_m_b': concat_dfs(data_dfs['birth_male']['bottom_m_b']),
    },

    'sixty_male': {
        'top_m_60': concat_dfs(data_dfs['sixty_male']['top_m_60']),
        'bottom_m_60': concat_dfs(data_dfs['sixty_male']['bottom_m_60']),
    },

    'birth_female': {
        'top_f_b': concat_dfs(data_dfs['birth_female']['top_f_b']),
        'bottom_f_b': concat_dfs(data_dfs['birth_female']['bottom_f_b']),
    },

    'sixty_female': {
        'top_f_60': concat_dfs(data_dfs['sixty_female']['top_f_60']),
        'bottom_f_60': concat_dfs(data_dfs['sixty_female']['bottom_f_60']),
    },

    'birth_both': {
        'top_b_b': concat_dfs(data_dfs['birth_both']['top_b_b']),
        'bottom_b_b': concat_dfs(data_dfs['birth_both']['bottom_b_b']),
    },

    'sixty_both': {
        'top_b_60': concat_dfs(data_dfs['sixty_both']['top_b_60']),
        'bottom_b_60': concat_dfs(data_dfs['sixty_both']['bottom_b_60']),
    },

    'rests_male': {
        'top_m_60': concat_dfs(data_dfs['rests_male']['top_m_60']),
        'bottom_m_60': concat_dfs(data_dfs['rests_male']['bottom_m_60']),
    },

    'rests_female': {
        'top_f_60': concat_dfs(data_dfs['rests_female']['top_f_60']),
        'bottom_f_60': concat_dfs(data_dfs['rests_female']['bottom_f_60']),
    },
    
    'rests_both': {
        'top_b_60': concat_dfs(data_dfs['rests_both']['top_b_60']),
        'bottom_b_60': concat_dfs(data_dfs['rests_both']['bottom_b_60'])
    }
}
```

##### Since comparing the absolute number of deaths across countries is not accurate, I am instead calculating the percentage distribution across the three major categories to allow for better comparisons between countries.

```python
def getting_percentage(variable):
    temp_variable = variable.groupby(['Country', 'Year'])['Value'].sum().reset_index()
    variable = pd.merge(temp_variable, variable, on=['Country', 'Year'], suffixes=('', '_original'))

# Divide the values in the two dataframes
    variable['Percentage'] = (variable['Value_original'] / variable['Value']) 

# Dropping the extra 'Value' column 
    variable = variable.drop(columns=['Value'])
    variable = variable.rename(columns={'Value_original': 'Value'})
    variable = variable.sort_values(by=['Year', 'Ranking'], ascending=True)
    return variable
```

```python
def type_after_60(df):
    # First, group the DataFrame by specified columns and aggregate the 'Value' column by summing
    variable = df.groupby(['Sex', 'Country', 'Country_code', 'Category', 'Sub_Category', 'Death_Cause', 'Year', 'Type', 'Ranking', 'categorize', 'Indicator'])['Value'].sum().reset_index()
    
    # Then, group by a subset of columns and aggregate the 'Value' column by taking the maximum value
    variable = variable.groupby(['Sex', 'Country', 'Country_code', 'Category', 'Sub_Category', 'Death_Cause', 'Year', 'Ranking', 'categorize', 'Indicator'])['Value'].max().reset_index()

    variable = variable.sort_values(by = (['Category', 'Year', 'Ranking']), ascending = True)
    return variable

# Create a dictionary of datasets to process, including both male, female, and both gender categories
type_results_dict = {
    'all_top_rests_male_Type': all_top_rests_male,
    'all_bottom_rests_male_Type': all_bottom_rests_male,
    'all_top_rests_female_Type': all_top_rests_female,
    'all_bottom_rests_female_Type': all_bottom_rests_female,
    'all_top_rests_both_Type': all_top_rests_both,
    'all_bottom_rests_both_Type': all_bottom_rests_both,
}

# Process each dataset using the 'type_after_60' function and update the dictionary with the results
for key, dataset in type_results_dict.items():
    type_results_dict[key] = type_after_60(dataset)

# Extract the processed datasets from the dictionary
all_top_rests_male_Type = type_results_dict['all_top_rests_male_Type']
all_bottom_rests_male_Type = type_results_dict['all_bottom_rests_male_Type']
all_top_rests_female_Type = type_results_dict['all_top_rests_female_Type']
all_bottom_rests_female_Type = type_results_dict['all_bottom_rests_female_Type']
all_top_rests_both_Type = type_results_dict['all_top_rests_both_Type']
all_bottom_rests_both_Type = type_results_dict['all_bottom_rests_both_Type']
```

##### Obtaining the percentage of each Death Type to get more info

```python
def getting_percentage_type(variable):
    # Grouping data by 'Country', 'Year', and 'Sub_Category' and summing 'Value'
    temp_variable = variable.groupby(['Country', 'Year', 'Sub_Category'])['Value'].sum().reset_index()
    
    # Merging the original data with the grouped data on 'Country', 'Year', and 'Sub_Category'
    variable = pd.merge(temp_variable, variable, on=['Country', 'Year', 'Sub_Category'], suffixes=('', '_original'))
    
    # Creating a mask to identify where 'Value' is non-zero
    mask = variable['Value'] != 0
    
    # Calculating 'Percentage' for rows where the mask is True, as the ratio of 'Value_original' to 'Value'
    variable.loc[mask, 'Percentage'] = variable.loc[mask, 'Value_original'] / variable.loc[mask, 'Value']
    
    # Assigning 0 to 'Percentage' where the mask is False
    variable.loc[~mask, 'Percentage'] = 0
    
    # Dropping the 'Value' column as it's no longer needed
    variable = variable.drop(columns=['Value'])
    
    # Renaming 'Value_original' back to 'Value' for consistency
    variable = variable.rename(columns={'Value_original': 'Value'})
    
    # Sorting the data by 'Year' and 'Ranking' in ascending order
    variable = variable.sort_values(by=['Year', 'Ranking'], ascending=True)
    
    return variable
```

```python
all_top_rests_male_Type = getting_percentage_type(all_top_rests_male_Type)
all_bottom_rests_male_Type = getting_percentage_type(all_bottom_rests_male_Type)
all_top_rests_female_Type = getting_percentage_type(all_top_rests_female_Type)
all_bottom_rests_female_Type = getting_percentage_type(all_bottom_rests_female_Type)
all_top_rests_both_Type = getting_percentage_type(all_top_rests_both_Type)
all_bottom_rests_both_Type = getting_percentage_type(all_bottom_rests_both_Type)
```
