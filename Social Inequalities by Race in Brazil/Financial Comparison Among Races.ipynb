```python
import pandas as pd
import numpy as np

# Importing the Data Frame from IBGE
file_path = 'C:\\Users\\Emerson Wu Li\\Desktop\\Files Para Social... Project\\Tabela 2.5 (Rend_Form_Geo).xls'
df_Financial = pd.read_excel(file_path, sheet_name = '2021')
df_cleaned = df_Financial.copy()

# Dropping all the Unnecessary rows (Description and Nan Values)
Rows_to_drop = [1, 2, 3, 66, 67, 68, 69, 70, 71, 72]
df_cleaned = df_cleaned.drop(Rows_to_drop)
df_cleaned = df_cleaned.dropna(how ='all')
df_cleaned
```

```python
# Reanaming the columns for better organization, easier to melt after
rename_dict = {
    'Unnamed: 2':'Branca',
    'Unnamed: 3':'Preta',
    'Unnamed: 4':'Parda',
    'Unnamed: 5':'Preta ou parda',
    'Unnamed: 7': 'Branca.1',
    'Unnamed: 8': 'Preta.1',
    'Unnamed: 9': 'Parda.1',
    'Unnamed: 10': 'Preta ou parda.1',
    'Unnamed: 12': 'Branca.2',
    'Unnamed: 13': 'Preta.2',
    'Unnamed: 14': 'Parda.2',
    'Unnamed: 15': 'Preta ou parda.2'
}

# Rename columns using the dictionary
df_cleaned = df_cleaned.rename(columns = rename_dict)

# Renaming the first column to 'r'(race)
df_cleaned.columns.values[0] = 'r'

# Dropping these columns because, is the total of the salary which I am not interested to use them
columns_to_drop = ['Unnamed: 1', 'Unnamed: 6', 'Unnamed: 11']
df_cleaned = df_cleaned.drop(columns = columns_to_drop)

# Again not necessary rows
Rows_to_drop = [4, 6]
df_cleaned = df_cleaned.drop(Rows_to_drop)
df_cleaned
```

```python
# Create a dictionary mapping Brazilian states and regions to their respective regions
Region = {
    'Norte': 'Norte', 'Rondônia': 'Norte', 'Acre': 'Norte', 'Amazonas': 'Norte', 'Roraima': 'Norte', 'Pará': 'Norte', 'Amapá': 'Norte', 'Tocantins': 'Norte',
    'Nordeste': 'Nordeste', 'Maranhão': 'Nordeste', 'Piauí': 'Nordeste', 'Ceará': 'Nordeste', 'Rio Grande do Norte': 'Nordeste', 'Paraíba': 'Nordeste', 'Pernambuco': 'Nordeste', 'Alagoas': 'Nordeste', 'Sergipe': 'Nordeste', 'Bahia': 'Nordeste',
    'Sudeste': 'Sudeste', 'Minas Gerais': 'Sudeste', 'Espírito Santo': 'Sudeste', 'Rio de Janeiro': 'Sudeste', 'São Paulo': 'Sudeste',
    'Sul': 'Sul', 'Paraná': 'Sul', 'Santa Catarina': 'Sul', 'Rio Grande do Sul': 'Sul',
    'Centro-Oeste': 'Centro-Oeste', 'Mato Grosso do Sul': 'Centro-Oeste', 'Mato Grosso': 'Centro-Oeste', 'Goiás': 'Centro-Oeste', 'Distrito Federal': 'Centro-Oeste'
}

# Create a new column 'Region' in the DataFrame 'df_cleaned'
# Use the 'apply' function to map the state names in column 'r'(race) to their respective regions based on the 'Region' dictionary
# If the state name is in the 'Region' dictionary, assign the corresponding region; otherwise, keep the state name as is
df_cleaned['Region'] = df_cleaned['r'].apply(lambda x: Region[x] if x in Region else x)

# Create a new column 'State' in the DataFrame 'df_cleaned'
# Use the 'apply' function to check if the state name in column 'r' is present in the 'Region' dictionary
# If the state name is in the dictionary, keep it; otherwise, assign an empty string
df_cleaned['State'] = df_cleaned['r'].apply(lambda x: x if x in Region else '')

# Drop the original column 'r' from the DataFrame 'df_cleaned' as it is no longer needed
df_cleaned = df_cleaned.drop(columns = 'r')

# These respective cities is not a Region so I dropped the rows that have them, 'Brasil' is the total of everything
region_to_drop = ['Brasil', 'Porto Velho', 'Rio Branco', 'Manaus', 'Boa Vista', 'Belém', 'Macapá', 'Palmas', 'São Luís', 'Teresina', 'Fortaleza', 'Natal', 'João Pessoa', 'Recife', 'Maceió', 'Aracaju', 'Salvador', 'Belo Horizonte', 'Vitória', 'Curitiba', 'Florianópolis', 'Porto Alegre', 'Campo Grande', 'Cuiabá', 'Goiânia', 'Brasília']
df_cleaned = df_cleaned[~df_cleaned['Region'].isin(region_to_drop)]
df_cleaned
```

```python
# Now I am separating into three section, because each section had a different type of job(Formal and Unofficial, Formal and Unofficial)
# So I created three Data Frame separating them
columns1 = ['Branca', 'Preta', 'Parda', 'Preta ou parda', 'Region', 'State']
columns2 = ['Branca.1', 'Preta.1', 'Parda.1', 'Preta ou parda.1', 'Region', 'State']
columns3 = ['Branca.2', 'Preta.2', 'Parda.2', 'Preta ou parda.2', 'Region', 'State']

df_Unofficial_Formal = df_cleaned[columns1].copy()
df_Formal = df_cleaned[columns2].copy()
df_Unofficial = df_cleaned[columns3].copy()

# Adding the column'Type' to specify which type the Data Frame is
df_Unofficial_Formal['Type'] = 'Formal and Unofficial'
df_Formal['Type'] = 'Formal'
df_Unofficial['Type'] = 'Unofficial'
```

```python
# Now renaming and translating for better understanding for 3 Data Frame
df_Formal = df_Formal.rename(columns={'Branca.1': 'Branca', 'Preta.1': 'Preta', 'Parda.1': 'Parda', 'Preta ou parda.1': 'Preta ou parda'})
df_Unofficial = df_Unofficial.rename(columns={'Branca.2': 'Branca', 'Preta.2': 'Preta', 'Parda.2': 'Parda', 'Preta ou parda.2': 'Preta ou parda'})

new_column_names = {
    'Branca': 'White',
    'Preta': 'Black',
    'Parda': 'Mixed',
    'Preta ou parda': 'Black or Mixed',
}

df_Unofficial_Formal = df_Unofficial_Formal.rename(columns = new_column_names)
df_Formal = df_Formal.rename(columns = new_column_names)
df_Unofficial = df_Unofficial.rename(columns = new_column_names)
```

```python
# Melting the DataFrames, by keeping the column 'Region', 'State' and 'Type' and transforming columns into row(with the column name 'Race')
# And the value name into 'Median_salary_per_month(R$)' 
df_Unofficial_Formal = df_Unofficial_Formal.melt(id_vars=['Region', 'State', 'Type'], 
                                                     value_vars=['White', 'Black', 'Mixed', 'Black or Mixed'],
                                                     var_name='Race', 
                                                     value_name='Median_salary_per_month(R$)')

df_Formal = df_Formal.melt(id_vars=['Region', 'State', 'Type'], 
                                 value_vars=['White', 'Black', 'Mixed', 'Black or Mixed'],
                                 var_name='Race', 
                                 value_name='Median_salary_per_month(R$)')

df_Unofficial = df_Unofficial.melt(id_vars=['Region', 'State', 'Type'], 
                                     value_vars=['White', 'Black', 'Mixed', 'Black or Mixed'],
                                     var_name='Race', 
                                     value_name='Median_salary_per_month(R$)')

# Concatenate all the melted DataFrames
df_result = pd.concat([df_Unofficial_Formal, df_Formal, df_Unofficial], ignore_index=True)
    
df_result['Year'] = 2021
df_result
```

```python
# Is going to be the same cleaning process, however there is multiple years to automate
# The code above was to test and check if the Data Frame was clean and ready to use

import pandas as pd
import numpy as np

def process_salary_data(file_path, sheet_name):
    df_Financial = pd.read_excel(file_path, sheet_name=sheet_name)

    df_cleaned = df_Financial.copy()
    Rows_to_drop = [1, 2, 3, 66, 67, 68, 69, 70, 71, 72]
    df_cleaned = df_cleaned.drop(Rows_to_drop)
    df_cleaned = df_cleaned.dropna(how='all')

    rename_dict = {
        'Unnamed: 2': 'Branca',
        'Unnamed: 3': 'Preta',
        'Unnamed: 4': 'Parda',
        'Unnamed: 5': 'Preta ou parda',
        'Unnamed: 7': 'Branca.1',
        'Unnamed: 8': 'Preta.1',
        'Unnamed: 9': 'Parda.1',
        'Unnamed: 10': 'Preta ou parda.1',
        'Unnamed: 12': 'Branca.2',
        'Unnamed: 13': 'Preta.2',
        'Unnamed: 14': 'Parda.2',
        'Unnamed: 15': 'Preta ou parda.2'
    }

    df_cleaned = df_cleaned.rename(columns=rename_dict)
    df_cleaned.columns.values[0] = 'r'
    columns_to_drop = ['Unnamed: 1', 'Unnamed: 6', 'Unnamed: 11']
    df_cleaned = df_cleaned.drop(columns=columns_to_drop)
    Rows_to_drop = [4, 6]
    df_cleaned = df_cleaned.drop(Rows_to_drop)

    Region = {
        'Norte': 'Norte', 'Rondônia': 'Norte', 'Acre': 'Norte', 'Amazonas': 'Norte', 'Roraima': 'Norte', 'Pará': 'Norte', 'Amapá': 'Norte', 'Tocantins': 'Norte',
        'Nordeste': 'Nordeste', 'Maranhão': 'Nordeste', 'Piauí': 'Nordeste', 'Ceará': 'Nordeste', 'Rio Grande do Norte': 'Nordeste', 'Paraíba': 'Nordeste', 'Pernambuco': 'Nordeste', 'Alagoas': 'Nordeste', 'Sergipe': 'Nordeste', 'Bahia': 'Nordeste',
        'Sudeste': 'Sudeste', 'Minas Gerais': 'Sudeste', 'Espírito Santo': 'Sudeste', 'Rio de Janeiro': 'Sudeste', 'São Paulo': 'Sudeste',
        'Sul': 'Sul', 'Paraná': 'Sul', 'Santa Catarina': 'Sul', 'Rio Grande do Sul': 'Sul',
        'Centro-Oeste': 'Centro-Oeste', 'Mato Grosso do Sul': 'Centro-Oeste', 'Mato Grosso': 'Centro-Oeste', 'Goiás': 'Centro-Oeste', 'Distrito Federal': 'Centro-Oeste'
    }

    df_cleaned.loc[:, 'Region'] = df_cleaned['r'].apply(lambda x: Region[x] if x in Region else x)
    df_cleaned.loc[:, 'State'] = df_cleaned['r'].apply(lambda x: x if x in Region else '')

    region_to_drop = ['Brasil', 'Porto Velho', 'Rio Branco', 'Manaus', 'Boa Vista', 'Belém', 'Macapá', 'Palmas', 'São Luís', 'Teresina', 'Fortaleza', 'Natal', 'João Pessoa', 'Recife', 'Maceió', 'Aracaju', 'Salvador', 'Belo Horizonte', 'Vitória', 'Curitiba', 'Florianópolis', 'Porto Alegre', 'Campo Grande', 'Cuiabá', 'Goiânia', 'Brasília']
    df_cleaned = df_cleaned[~df_cleaned['Region'].isin(region_to_drop)]
    df_cleaned = df_cleaned.drop(columns='r')

    columns1 = ['Branca', 'Preta', 'Parda', 'Preta ou parda', 'Region', 'State']
    columns2 = ['Branca.1', 'Preta.1', 'Parda.1', 'Preta ou parda.1', 'Region', 'State']
    columns3 = ['Branca.2', 'Preta.2', 'Parda.2', 'Preta ou parda.2', 'Region', 'State']

    df_Unofficial_Formal = df_cleaned[columns1].copy()
    df_Formal = df_cleaned[columns2].copy()
    df_Unofficial = df_cleaned[columns3].copy()

    df_Unofficial_Formal['Type'] = 'Formal and Unofficial'
    df_Formal['Type'] = 'Formal'
    df_Unofficial['Type'] = 'Unofficial'

    df_Formal = df_Formal.rename(columns={'Branca.1': 'Branca', 'Preta.1': 'Preta', 'Parda.1': 'Parda', 'Preta ou parda.1': 'Preta ou parda'})
    df_Unofficial = df_Unofficial.rename(columns={'Branca.2': 'Branca', 'Preta.2': 'Preta', 'Parda.2': 'Parda', 'Preta ou parda.2': 'Preta ou parda'})

    new_column_names = {
        'Branca': 'White',
        'Preta': 'Black',
        'Parda': 'Mixed',
        'Preta ou parda': 'Black or Mixed',
    }

    df_Unofficial_Formal = df_Unofficial_Formal.rename(columns=new_column_names)
    df_Formal = df_Formal.rename(columns=new_column_names)
    df_Unofficial = df_Unofficial.rename(columns=new_column_names)

    df_Unofficial_Formal = df_Unofficial_Formal.melt(id_vars=['Region', 'State', 'Type'], 
                                                     value_vars=['White', 'Black', 'Mixed', 'Black or Mixed'],
                                                     var_name='Race', 
                                                     value_name='Median_salary_per_month(R$)')

    df_Formal = df_Formal.melt(id_vars=['Region', 'State', 'Type'], 
                                 value_vars=['White', 'Black', 'Mixed', 'Black or Mixed'],
                                 var_name='Race', 
                                 value_name='Median_salary_per_month(R$)')

    df_Unofficial = df_Unofficial.melt(id_vars=['Region', 'State', 'Type'], 
                                     value_vars=['White', 'Black', 'Mixed', 'Black or Mixed'],
                                     var_name='Race', 
                                     value_name='Median_salary_per_month(R$)')

    df_result = pd.concat([df_Unofficial_Formal, df_Formal, df_Unofficial], ignore_index=True)

    return df_result

def combine_salary_data(file_path, years):
    combined_df = pd.DataFrame()
    
    for year in years:
        sheet_name = str(year)
        df_year = process_salary_data(file_path, sheet_name)
        df_year['Year'] = year
        combined_df = pd.concat([combined_df, df_year], ignore_index=True)
    
    return combined_df

file_path = 'C:\\Users\\Emerson Wu Li\\Desktop\\Files Para Social... Project\\Tabela 2.5 (Rend_Form_Geo).xls'
years = [2021, 2020, 2019, 2018, 2017, 2016, 2015, 2014, 2013, 2012]

df_Finantial = combine_salary_data(file_path, years)
df_Finantial
```

```python
# Filtering Type for Formal
df_Salary_Race_Region_Formal = df_Finantial.copy()
df_Salary_Race_Region_Formal = df_Salary_Race_Region_Formal[df_Salary_Race_Region_Formal['Type'] == 'Formal']

# Filtering Race
race_in = ['White', 'Black', 'Mixed']
df_Salary_Race_Region_Formal = df_Salary_Race_Region_Formal[df_Salary_Race_Region_Formal['Race'].isin(race_in)]

# Keeping the necessary region and dropping the state column
State_to_keep = ['Norte', 'Nordeste', 'Sudeste', 'Sul', 'Centro-Oeste']
df_Salary_Race_Region_Formal = df_Salary_Race_Region_Formal[df_Salary_Race_Region_Formal['State'].isin(State_to_keep)]
df_Salary_Race_Region_Formal = df_Salary_Race_Region_Formal.drop(columns = 'State')

# Now I only want the year 2021
df_Salary_Race_Region_Formal_2021 = df_Salary_Race_Region_Formal[df_Salary_Race_Region_Formal['Year'] == 2021]

df_Salary_Race_Region_Formal_2021
```

```python
# Making a copy of the previous Data Frame
df_Salary_Race_Region_Formal_race = df_Salary_Race_Region_Formal.copy()
df_Salary_Race_Region_Formal_race = df_Salary_Race_Region_Formal_race.drop(columns = 'Region')

# Grouping by Year, Race and type and aggregating Salary to get the median of each year
df_Salary_Race_Region_Formal_race = df_Salary_Race_Region_Formal_race.groupby(['Year', 'Race', 'Type'])['Median_salary_per_month(R$)'].agg('median').reset_index()

# Sorting the values by year
df_Salary_Race_Region_Formal_race = df_Salary_Race_Region_Formal_race.sort_values(by = 'Year', ascending = False)
df_Salary_Race_Region_Formal_race
```
