```python
# Importing the Data Frame from IBGE
import pandas as pd
import numpy as np


file_path = 'C:\\Users\\Emerson Wu Li\\Desktop\\Files Para Social... Project\\Tabela 3.1 (Pop_DomProp).xls'
df_Homelessness = pd.read_excel(file_path, sheet_name = '2019')
df_cleaned = df_Homelessness.copy()
df_cleaned
```

```python
# Dropping all the NaN row values
df_cleaned = df_cleaned.dropna(how = 'all')

# Setting the specified row to be the column
df_cleaned.columns = df_cleaned.iloc[3]

# Dropping the unnecessary rows such as description
rows_to_drop = [2, 3, 4, 5, 40]
df_cleaned = df_cleaned.drop(rows_to_drop)
df_cleaned   
```

```python
# Changing first three column names
df_cleaned.columns.values[0] = 'r'
df_cleaned.columns.values[1] = 'Quantity(10^3)'
df_cleaned.columns.values[2] = 'Total'

# Translating to english for better understanding
new_column_names = {
    'Branca': 'White',
    'Preta': 'Black',
    'Parda': 'Mixed',
    'Preta ou parda': 'Black or Mixed',
    'Quantity': 'Quantity(10^3)'
}

df_cleaned = df_cleaned.rename(columns = new_column_names)
df_cleaned
```

```python
# Create a dictionary mapping Brazilian states and regions to their respective regions
Region = {
    'Norte': 'Norte', 'Rondônia': 'Norte', 'Acre': 'Norte', 'Amazonas': 'Norte', 'Roraima': 'Norte', 'Pará': 'Norte', 'Amapá': 'Norte', 'Tocantins': 'Norte',
    'Nordeste': 'Nordeste', 'Maranhão': 'Nordeste', 'Piauí': 'Nordeste', 'Ceará': 'Nordeste', 'Rio Grande do Norte': 'Nordeste', 'Paraíba': 'Nordeste', 'Pernambuco': 'Nordeste', 'Alagoas': 'Nordeste', 'Sergipe': 'Nordeste', 'Bahia': 'Nordeste',
    'Sudeste': 'Sudeste', 'Minas Gerais': 'Sudeste', 'Espírito Santo': 'Sudeste', 'Rio de Janeiro': 'Sudeste', 'São Paulo': 'Sudeste',
    'Sul': 'Sul', 'Paraná': 'Sul', 'Santa Catarina': 'Sul', 'Rio Grande do Sul': 'Sul',
    'Centro-Oeste': 'Centro-Oeste', 'Mato Grosso do Sul': 'Centro-Oeste', 'Mato Grosso': 'Centro-Oeste', 'Goiás': 'Centro-Oeste', 'Distrito Federal': 'Centro-Oeste'
}

# Create a new column 'Region' in the DataFrame 'df_cleaned'
# Use the 'apply' function to map the state names in column 'r'(region) to their respective regions based on the 'Region' dictionary
# If the state name is in the 'Region' dictionary, assign the corresponding region; otherwise, keep the state name as is
df_cleaned['Region'] = df_cleaned['r'].apply(lambda x: Region[x] if x in Region else x)

# Create a new column 'State' in the DataFrame 'df_cleaned'
# Use the 'apply' function to check if the state name in column 'r' is present in the 'Region' dictionary
# If the state name is in the dictionary, keep it; otherwise, assign an empty string
df_cleaned['State'] = df_cleaned['r'].apply(lambda x: x if x in Region else '')

# Drop the original column 'r' from the DataFrame 'df_cleaned' as it is no longer needed
df_cleaned = df_cleaned.drop(columns = 'r')

df_cleaned
```

```python
# Melting the Data Frame to transform the columns into rows(Race), and keeping the Region, State, and Quantity(10^3)

df_cleaned = df_cleaned.melt(id_vars=['Region', 'State', 'Quantity(10^3)'], 
                value_vars = ['White', 'Black', 'Mixed', 'Black or Mixed'],
                var_name = 'Race',  
                value_name ='Proportion_of_not_homeless')

df_cleaned
```

```python
# I wanted to insert a new column containing the Proportion of homeless people
df_cleaned = df_cleaned.drop(columns = 'Quantity(10^3)')
df_cleaned['Proportion_of_homeless'] = 100 - df_cleaned['Proportion_of_not_homeless']

# Adding the column Year to specify which year it is and dropping the total proportion (Brazil)
df_cleaned = df_cleaned.drop(0)
df_cleaned['Year'] = 2019
df_cleaned
```

```python
# Is going to be the same cleaning process, however there is multiple years to automate
# The code above was to test and check if the Data Frame was clean and ready to use

def process_population_data_homeless(file_path, sheet_name):

    df_Homelessness = pd.read_excel(file_path, sheet_name = sheet_name)


    df_cleaned = df_Homelessness.copy()
    df_cleaned = df_cleaned.dropna(how = 'all')
    df_cleaned.columns = df_cleaned.iloc[3]

    rows_to_drop = [2, 3, 4, 5, 40]
    df_cleaned = df_cleaned.drop(rows_to_drop)
    df_cleaned.columns.values[0] = 'r'
    df_cleaned.columns.values[1] = 'Quantity(10^3)'
    df_cleaned.columns.values[2] = 'Total'


    new_column_names = {
        'Branca': 'White',
        'Preta': 'Black',
        'Parda': 'Mixed',
        'Preta ou parda': 'Black or Mixed',
        'Quantity': 'Quantity(10^3)'
    }

    df_cleaned = df_cleaned.rename(columns = new_column_names)

    Region = {
    'Norte': 'Norte', 'Rondônia': 'Norte', 'Acre': 'Norte', 'Amazonas': 'Norte', 'Roraima': 'Norte', 'Pará': 'Norte', 'Amapá': 'Norte', 'Tocantins': 'Norte',
    'Nordeste': 'Nordeste', 'Maranhão': 'Nordeste', 'Piauí': 'Nordeste', 'Ceará': 'Nordeste', 'Rio Grande do Norte': 'Nordeste', 'Paraíba': 'Nordeste', 'Pernambuco': 'Nordeste', 'Alagoas': 'Nordeste', 'Sergipe': 'Nordeste', 'Bahia': 'Nordeste',
    'Sudeste': 'Sudeste', 'Minas Gerais': 'Sudeste', 'Espírito Santo': 'Sudeste', 'Rio de Janeiro': 'Sudeste', 'São Paulo': 'Sudeste',
    'Sul': 'Sul', 'Paraná': 'Sul', 'Santa Catarina': 'Sul', 'Rio Grande do Sul': 'Sul',
    'Centro-Oeste': 'Centro-Oeste', 'Mato Grosso do Sul': 'Centro-Oeste', 'Mato Grosso': 'Centro-Oeste', 'Goiás': 'Centro-Oeste', 'Distrito Federal': 'Centro-Oeste'
    }


    df_cleaned['Region'] = df_cleaned['r'].apply(lambda x: Region[x] if x in Region else x)
    df_cleaned['State'] = df_cleaned['r'].apply(lambda x: x if x in Region else '')
    df_cleaned = df_cleaned.drop(columns = 'r')
    

    df_cleaned = df_cleaned.melt(id_vars=['Region', 'State', 'Quantity(10^3)'], 
                    value_vars = ['White', 'Black', 'Mixed', 'Black or Mixed'],
                    var_name = 'Race',  
                    value_name ='Proportion_of_not_homeless')
    
    return df_cleaned

def combine_data(file_path, years):
    combined_df = pd.DataFrame()

    for year in years:
        sheet_name = str(year)
        df_year = process_population_data_homeless(file_path, sheet_name)
        df_year['Year'] = year
        combined_df = pd.concat([combined_df, df_year], ignore_index = True)
    
    return combined_df

file_path = 'C:\\Users\\Emerson Wu Li\\Desktop\\Files Para Social... Project\\Tabela 3.1 (Pop_DomProp).xls'
years = [2019, 2018, 2017, 2016] 

df_Homelessness = combine_data(file_path, years)
df_Homelessness = df_Homelessness.drop(columns = 'Quantity(10^3)')
df_Homelessness['Proportion_of_homeless'] = 100 - df_Homelessness['Proportion_of_not_homeless']
df_Homelessness = df_Homelessness.drop(0)
df_Homelessness
```

```python
# In this analysis I only want the Region and in the year 2019
df_Homelessness_only_region = df_Homelessness.copy()

# Filtering for year 2019
df_Homelessness_only_region = df_Homelessness_only_region[df_Homelessness_only_region['Year'] == 2019]

# Keeping the races that I need
row_to_keep = ['White', 'Black', 'Mixed']
df_Homelessness_only_region = df_Homelessness_only_region[df_Homelessness_only_region['Race'].isin(row_to_keep)]

# Grouping by region, race and year, then calculating the median of homeless people in each region by race
df_Homelessness_only_region = df_Homelessness_only_region.groupby(['Region', 'Race', 'Year'])['Proportion_of_homeless'].agg('median').reset_index()

# Dropping the rows that contains total proportion
df_Homelessness_only_region = df_Homelessness_only_region.drop([0, 1])
df_Homelessness_only_region
```

```python
# In this analysis I want to display the Proportion of homeless people by race
df_Homelessness_only_race = df_Homelessness.copy()

# Dropping the columns that I dont need
columns_to_drop = ['Region', 'Proportion_of_not_homeless']
df_Homelessness_only_race = df_Homelessness_only_race.drop(columns = columns_to_drop)

# Grouping by Race and Year, and calculating the median of homeless people
df_Homelessness_only_race = df_Homelessness_only_race.groupby(['Race', 'Year'])['Proportion_of_homeless'].agg('median').reset_index()
df_Homelessness_only_race = df_Homelessness_only_race.sort_values(by = 'Year', ascending = False)

df_Homelessness_only_race
```
