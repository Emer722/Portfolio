```python
import pandas as pd
import numpy as np

# Importing the Data Frame from IBGE
file_path = 'C:\\Users\\Emerson Wu Li\\Desktop\\Files Para Social... Project\\Tabela 1.1 (Pop_Geo).xls'
df_Demographic_Distribution = pd.read_excel(file_path, sheet_name = '2021')
df_cleaned = df_Demographic_Distribution.copy()
df_cleaned

```

```python
# By dropping all the NaN rows values
df_cleaned = df_cleaned.dropna(how = 'all')
# Setting the 4 row to be the column name, it will be easier to melt
df_cleaned.columns = df_cleaned.iloc[2]
df_cleaned
```

```python
# Droping all the rows that is not necessary(NaN or description)
rows_to_drop = [2, 3, 4, 5, 39]
df_cleaned = df_cleaned.drop(rows_to_drop)

# Renaming the columns name
df_cleaned.columns.values[1] = 'Quantity'
df_cleaned.columns.values[0] = 'r'

# Tranalating to English for better understanding 
new_column_names = {
    'Branca': 'White',
    'Preta': 'Black',
    'Parda': 'Mixed',
    'Preta ou parda': 'Black or Mixed',
    'Quantity': 'Quantity1(10^3)'
}

df_cleaned = df_cleaned.rename(columns = new_column_names)
df_cleaned
```

```python
# Create a dictionary mapping Brazilian states and regions to their respective regions
Region = {
    'Norte': 'Norte', 'Rondônia': 'Norte', 'Acre': 'Norte', 'Amazonas': 'Norte', 'Roraima': 'Norte', 'Pará': 'Norte', 'Amapá': 'Norte', 'Tocantins': 'Norte',
    'Nordeste': 'Nordeste', 'Maranhão': 'Nordeste', 'Piauí': 'Nordeste', 'Ceará': 'Nordeste', 'Rio Grande do Norte': 'Nordeste', 'Paraíba': 'Nordeste', 'Pernambuco': 'Nordeste', 'Alagoas': 'Nordeste', 'Sergipe': 'Nordeste', 'Bahia': 'Nordeste',
    'Sudeste': 'Sudeste', 'Minas Gerais': 'Sudeste', 'Espírito Santo': 'Sudeste', 'Rio de Janeiro': 'Sudeste', 'São Paulo': 'Sudeste',
    'Sul': 'Sul', 'Paraná': 'Sul', 'Santa Catarina': 'Sul', 'Rio Grande do Sul': 'Sul',
    'Centro-Oeste': 'Centro-Oeste', 'Mato Grosso do Sul': 'Centro-Oeste', 'Mato Grosso': 'Centro-Oeste', 'Goiás': 'Centro-Oeste', 'Distrito Federal': 'Centro-Oeste'
}

# Create a new column 'Region' in the DataFrame 'df_cleaned'
# Use the 'apply' function to map the state names in column 'r'(region) to their respective regions based on the 'Region' dictionary
# If the state name is in the 'Region' dictionary, assign the corresponding region; otherwise, keep the state name as is
df_cleaned['Region'] = df_cleaned['r'].apply(lambda x: Region[x] if x in Region else x)

# Create a new column 'State' in the DataFrame 'df_cleaned'
# Use the 'apply' function to check if the state name in column 'r' is present in the 'Region' dictionary
# If the state name is in the dictionary, keep it; otherwise, assign an empty string
df_cleaned['State'] = df_cleaned['r'].apply(lambda x: x if x in Region else '')

# Drop the original column 'r' from the DataFrame 'df_cleaned' as it is no longer needed
df_cleaned = df_cleaned.drop(columns = 'r')
df_cleaned
```

```python
# To transform Columns into row, by keeping the columns(Region, State, and Quantity1(10^3)
# And the rows values to be (White, Black, Mixed and Black ot Mixed) in the Race column to their respective percentage
df_cleaned = df_cleaned.melt(id_vars=['Region', 'State', 'Quantity1(10^3)'], 
                    value_vars = ['White', 'Black', 'Mixed', 'Black or Mixed'],
                    var_name = 'Race',  
                    value_name = 'Percentage')

# Creating a Year column   
df_cleaned['Year'] = 2021
df_cleaned
```

```python
# The percentage of the population was based on region and state, therefore I got the actual quantity for each race and State
df_cleaned['Quantity(10^3)'] = (df_cleaned['Quantity1(10^3)'] * df_cleaned['Percentage']) / 100

columns_to_drop = ['Quantity1(10^3)', 'Percentage']
df_cleaned = df_cleaned.drop(columns = columns_to_drop)

# Rounding the quantity
df_cleaned['Quantity(10^3)'] = df_cleaned['Quantity(10^3)'].round().astype(int)

# Translating into english for better understanding
translation_dict = {
        'Norte': 'North',
        'Nordeste': 'Northeast',
        'Sudeste': 'Southeast',
        'Sul': 'South',
        'Centro-Oeste': 'Central-West'
    }

# Replace Portuguese region names with English names in the 'Region' column
df_cleaned['Region'] = df_cleaned['Region'].replace(translation_dict)

df_cleaned
```

```python
# Is going to be the same cleaning process, however there is multiple years to automate
# The code above was to test and check if the Data Frame was clean and ready to use
def process_population_data(file_path, sheet_name):

    # Read the Excel file and load the specified sheet into a DataFrame
    df_Demographic_Distribution = pd.read_excel(file_path, sheet_name = sheet_name)
    df_cleaned = df_Demographic_Distribution.copy()
    df_cleaned = df_cleaned.dropna(how = 'all')
    df_cleaned.columns = df_cleaned.iloc[2]

    rows_to_drop = [2, 3, 4, 5, 39]
    df_cleaned = df_cleaned.drop(rows_to_drop)
    df_cleaned.columns.values[1] = 'Quantity'
    df_cleaned.columns.values[0] = 'r'

    new_column_names = {
        'Branca': 'White',
        'Preta': 'Black',
        'Parda': 'Mixed',
        'Preta ou parda': 'Black or Mixed',
        'Quantity': 'Quantity1(10^3)'
    }

    df_cleaned = df_cleaned.rename(columns = new_column_names)



    Region = {
        'Norte': 'Norte', 'Rondônia': 'Norte', 'Acre': 'Norte', 'Amazonas': 'Norte', 'Roraima': 'Norte', 'Pará': 'Norte', 'Amapá': 'Norte', 'Tocantins': 'Norte',
        'Nordeste': 'Nordeste', 'Maranhão': 'Nordeste', 'Piauí': 'Nordeste', 'Ceará': 'Nordeste', 'Rio Grande do Norte': 'Nordeste', 'Paraíba': 'Nordeste', 'Pernambuco': 'Nordeste', 'Alagoas': 'Nordeste', 'Sergipe': 'Nordeste', 'Bahia': 'Nordeste',
        'Sudeste': 'Sudeste', 'Minas Gerais': 'Sudeste', 'Espírito Santo': 'Sudeste', 'Rio de Janeiro': 'Sudeste', 'São Paulo': 'Sudeste',
        'Sul': 'Sul', 'Paraná': 'Sul', 'Santa Catarina': 'Sul', 'Rio Grande do Sul': 'Sul',
        'Centro-Oeste': 'Centro-Oeste', 'Mato Grosso do Sul': 'Centro-Oeste', 'Mato Grosso': 'Centro-Oeste', 'Goiás': 'Centro-Oeste', 'Distrito Federal': 'Centro-Oeste'
    }


    df_cleaned['Region'] = df_cleaned['r'].apply(lambda x: Region[x] if x in Region else x)
    df_cleaned['State'] = df_cleaned['r'].apply(lambda x: x if x in Region else '')
    df_cleaned = df_cleaned.drop(columns = 'r')


    df_cleaned = df_cleaned.melt(id_vars=['Region', 'State', 'Quantity1(10^3)'], 
                    value_vars = ['White', 'Black', 'Mixed', 'Black or Mixed'],
                    var_name = 'Race',  
                    value_name ='Percentage')
    
    # Return the reshaped DataFrame
    return df_cleaned

    

def combine_data(file_path, years):

    # Initialize an empty DataFrame to store combined data
    combined_df = pd.DataFrame()

    # Iterate over each year provided in the 'years' list
    for year in years:

        # Convert year to string to match sheet names
        sheet_name = str(year)

        # Process the data for the given year
        df_year = process_population_data(file_path, sheet_name)

        # Add a new column 'Year' to the DataFrame
        df_year['Year'] = year

        # Concatenate the current year's data with the combined DataFrame
        combined_df = pd.concat([combined_df, df_year], ignore_index=True)
        
    # Return the combined DataFrame containing data for all specified years
    return combined_df


file_path = 'C:\\Users\\Emerson Wu Li\\Desktop\\Files Para Social... Project\\Tabela 1.1 (Pop_Geo).xls'
years = [2021, 2020, 2019, 2018, 2017, 2016, 2015, 2014, 2013, 2012]
df_cleaned = combine_data(file_path, years)


df_cleaned['Quantity(10^3)'] = (df_cleaned['Quantity1(10^3)'] * df_cleaned['Percentage']) / 100

columns_to_drop = ['Quantity1(10^3)', 'Percentage']
df_cleaned = df_cleaned.drop(columns = columns_to_drop)


df_cleaned['Quantity(10^3)'] = df_cleaned['Quantity(10^3)'].round().astype(int)

translation_dict = {
        'Norte': 'North',
        'Nordeste': 'Northeast',
        'Sudeste': 'Southeast',
        'Sul': 'South',
        'Centro-Oeste': 'Central-West'
    }


df_cleaned['Region'] = df_cleaned['Region'].replace(translation_dict)
df_cleaned
```

```python
# Getting the percentage of race in each region
# Filtering for State and Race
df_Distribution_only_region = df_cleaned.copy()

# Specifying that I dont want the states
State_to_keep = ['Norte', 'Nordeste', 'Sudeste', 'Sul', 'Centro-Oeste']
df_Distribution_only_region = df_Distribution_only_region[df_Distribution_only_region['State'].isin(State_to_keep)] 

# Eliminating 'Black and Mixed'
race_that_want = ['White', 'Black', 'Mixed']
df_Distribution_only_region = df_Distribution_only_region[df_Distribution_only_region['Race'].isin(race_that_want)]

df_Distribution_only_region
```

```python
# Calculating the percentage by race in each region

# Grouping by Region and Year and summing them
group = df_Distribution_only_region.groupby(['Region', 'Year'])['Quantity(10^3)'].transform('sum')

# Inserting the result from the variable 'group' into a new column in the Data Frame
df_Distribution_only_region['Total'] = group
df_Distribution_only_region
```

```python
# Now dividing the quantity of each race to the 'total' column, to obtaing the actual percentage
# For example the north region has 17% white people and in northeast has 25% of them and so on.
df_Distribution_only_region['Percentage'] = (df_Distribution_only_region['Quantity(10^3)'] / df_Distribution_only_region['Total']) * 100
df_Distribution_only_region = df_Distribution_only_region.drop(columns = ['State', 'Total'])
df_Distribution_only_region
```
