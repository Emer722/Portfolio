```python
# Is going to be the same cleaning process, however there is multiple years to automate
# The code above was to test and check if the Data Frame was clean and ready to use

def process_population_data_homicide(file_path, sheet_name):
    df = pd.read_excel(file_path, sheet_name=sheet_name)

    df_cleaned = df.copy()
    df_cleaned = df_cleaned.dropna(how='all')
    rows_to_drop = [1, 2, 24, 25]
    df_cleaned = df_cleaned.drop(rows_to_drop)
    df_cleaned = df_cleaned.iloc[4:23].reset_index(drop=True)

    columns = ['Gender', 'Total_Absoluto', 'Total_Taxa', '0-14_Absoluto', '0-14_Taxa', 
               '15-29_Absoluto', '15-29_Taxa', '30-59_Absoluto', '30-59_Taxa', 
               '60+_Absoluto', '60+_Taxa']
    df_cleaned.columns = columns

    rows_to_drop = [2, 7]
    df_cleaned = df_cleaned.drop(rows_to_drop)

    columns1 = ['Gender', 'Total_Absoluto', '0-14_Absoluto', '15-29_Absoluto', '30-59_Absoluto', '60+_Absoluto']
    columns2 = ['Gender', 'Total_Taxa', '0-14_Taxa', '15-29_Taxa', '30-59_Taxa', '60+_Taxa']

    df_cleaned_1 = df_cleaned[columns1]
    df_cleaned_2 = df_cleaned[columns2]

    melted_df1 = df_cleaned_1.melt(id_vars=['Gender'], 
                                   value_vars=columns1[1:],
                                   var_name='Age_Group_Number', 
                                   value_name='Quantity')

    melted_df2 = df_cleaned_2.melt(id_vars=['Gender'], 
                                   value_vars=columns2[1:],
                                   var_name='Age_Group_Percentage', 
                                   value_name='Percentage')

    melted_df1['common'] = melted_df1['Gender'] + melted_df1['Age_Group_Number'].str.split('_').str[0]
    melted_df2['common'] = melted_df2['Gender'] + melted_df2['Age_Group_Percentage'].str.split('_').str[0]

    df_combined = pd.merge(melted_df1, melted_df2, on='common')
    df_combined = df_combined[['Gender_x', 'Age_Group_Number', 'Quantity', 'Age_Group_Percentage', 'Percentage']]
    df_combined.columns = ['Gender', 'Age_Group_Number', 'Quantity', 'Age_Group_Percentage', 'Percentage']
    df_combined = df_combined.drop(columns='Age_Group_Percentage')
    df_combined = df_combined.rename(columns={'Age_Group_Number': 'Age_Group'})
    df_combined['Age_Group'] = df_combined['Age_Group'].str.replace('_Absoluto', '')

    return df_combined

def combine_data(file_path, years):
    combined_df = pd.DataFrame()

    for year in years:
        sheet_name = str(year)
        df_year = process_population_data_homicide(file_path, sheet_name)
        df_year['Year'] = year
        combined_df = pd.concat([combined_df, df_year], ignore_index=True)
    
    return combined_df


file_path = 'C:\\Users\\Emerson Wu Li\\Desktop\\Files Para Social... Project\\Tabela 5.5 (Tx_Homic_Idad).xls'
years = [2020, 2019, 2018, 2017, 2016, 2015, 2014, 2013, 2012] 

df_combined = combine_data(file_path, years)

replacement_dict = {
    'Homem preto ou pardo': 'Black or Mixed Men',
    'Mulher preta ou parda': 'Black or Mixed Women',
    'Mulher preta': 'Black Women',
    'Mulher parda': 'Mixed Women',
    'Mulher branca': 'White Women',
    'Homem': 'Men',
    'Mulher': 'Women',
    'Branca': 'White',
    'Preta': 'Black',
    'Parda': 'Mixed',
    'preto': 'Black',
    'pardo': 'Mixed',
    'branco': 'White'
}

df_combined['Gender'] = df_combined['Gender'].replace(replacement_dict, regex=True)
df_combined = df_combined.rename(columns={'Gender': 'Gender and Race', 'Percentage': 'Homicide Rate'})
df_combined['Year'] = df_combined['Year'].astype(int)  

df_Homicide = df_combined.copy()
df_Homicide
```

```python
# In this analysis I want to see the quantity of homicide cases by race over the years

df_Homicide_only_race = df_Homicide.copy()

# Filtering the Race
race_to_keep = ['White', 'Black', 'Mixed']
df_Homicide_only_race = df_Homicide_only_race[df_Homicide_only_race['Gender and Race'].isin(race_to_keep)]

# Filtering for the Age Group
age_group_to_keep = ['Total']
df_Homicide_only_race = df_Homicide_only_race[df_Homicide_only_race['Age_Group'].isin(age_group_to_keep)]

#  Dropping the unwanted columns
columns_to_drop = ['Age_Group', 'Homicide Rate']
df_Homicide_only_race = df_Homicide_only_race.drop(columns = columns_to_drop)
df_Homicide_only_race = df_Homicide_only_race.rename(columns = {'Gender and Race' : 'Race'})
df_Homicide_only_race
```

```python
# In this analysis I want to check if the Homicide Rate is correct or not
# So I need to bring back the first table, that contains the population quantity
def process_population_data(file_path, sheet_name):

    # Read the Excel file and load the specified sheet into a DataFrame
    df_Demographic_Distribution = pd.read_excel(file_path, sheet_name = sheet_name)
    df_cleaned = df_Demographic_Distribution.copy()
    df_cleaned = df_cleaned.dropna(how = 'all')
    df_cleaned.columns = df_cleaned.iloc[2]

    rows_to_drop = [2, 3, 4, 5, 39]
    df_cleaned = df_cleaned.drop(rows_to_drop)
    df_cleaned.columns.values[1] = 'Quantity'
    df_cleaned.columns.values[0] = 'r'

    new_column_names = {
        'Branca': 'White',
        'Preta': 'Black',
        'Parda': 'Mixed',
        'Preta ou parda': 'Black or Mixed',
        'Quantity': 'Quantity1(10^3)'
    }

    df_cleaned = df_cleaned.rename(columns = new_column_names)



    Region = {
        'Norte': 'Norte', 'Rondônia': 'Norte', 'Acre': 'Norte', 'Amazonas': 'Norte', 'Roraima': 'Norte', 'Pará': 'Norte', 'Amapá': 'Norte', 'Tocantins': 'Norte',
        'Nordeste': 'Nordeste', 'Maranhão': 'Nordeste', 'Piauí': 'Nordeste', 'Ceará': 'Nordeste', 'Rio Grande do Norte': 'Nordeste', 'Paraíba': 'Nordeste', 'Pernambuco': 'Nordeste', 'Alagoas': 'Nordeste', 'Sergipe': 'Nordeste', 'Bahia': 'Nordeste',
        'Sudeste': 'Sudeste', 'Minas Gerais': 'Sudeste', 'Espírito Santo': 'Sudeste', 'Rio de Janeiro': 'Sudeste', 'São Paulo': 'Sudeste',
        'Sul': 'Sul', 'Paraná': 'Sul', 'Santa Catarina': 'Sul', 'Rio Grande do Sul': 'Sul',
        'Centro-Oeste': 'Centro-Oeste', 'Mato Grosso do Sul': 'Centro-Oeste', 'Mato Grosso': 'Centro-Oeste', 'Goiás': 'Centro-Oeste', 'Distrito Federal': 'Centro-Oeste'
    }


    df_cleaned['Region'] = df_cleaned['r'].apply(lambda x: Region[x] if x in Region else x)
    df_cleaned['State'] = df_cleaned['r'].apply(lambda x: x if x in Region else '')
    df_cleaned = df_cleaned.drop(columns = 'r')


    df_cleaned = df_cleaned.melt(id_vars=['Region', 'State', 'Quantity1(10^3)'], 
                    value_vars = ['White', 'Black', 'Mixed', 'Black or Mixed'],
                    var_name = 'Race',  
                    value_name ='Percentage')
    
    # Return the reshaped DataFrame
    return df_cleaned

    

def combine_data(file_path, years):

    # Initialize an empty DataFrame to store combined data
    combined_df = pd.DataFrame()

    # Iterate over each year provided in the 'years' list
    for year in years:

        # Convert year to string to match sheet names
        sheet_name = str(year)

        # Process the data for the given year
        df_year = process_population_data(file_path, sheet_name)

        # Add a new column 'Year' to the DataFrame
        df_year['Year'] = year

        # Concatenate the current year's data with the combined DataFrame
        combined_df = pd.concat([combined_df, df_year], ignore_index=True)
        
    # Return the combined DataFrame containing data for all specified years
    return combined_df


file_path = 'C:\\Users\\Emerson Wu Li\\Desktop\\Files Para Social... Project\\Tabela 1.1 (Pop_Geo).xls'
years = [2021, 2020, 2019, 2018, 2017, 2016, 2015, 2014, 2013, 2012]
df_cleaned = combine_data(file_path, years)


df_cleaned['Quantity(10^3)'] = (df_cleaned['Quantity1(10^3)'] * df_cleaned['Percentage']) / 100

columns_to_drop = ['Quantity1(10^3)', 'Percentage']
df_cleaned = df_cleaned.drop(columns = columns_to_drop)


df_cleaned['Quantity(10^3)'] = df_cleaned['Quantity(10^3)'].round().astype(int)

translation_dict = {
        'Norte': 'North',
        'Nordeste': 'Northeast',
        'Sudeste': 'Southeast',
        'Sul': 'South',
        'Centro-Oeste': 'Central-West'
    }


df_cleaned['Region'] = df_cleaned['Region'].replace(translation_dict)

df_Distribution_only_region = df_cleaned.copy()

State_to_keep = ['Norte', 'Nordeste', 'Sudeste', 'Sul', 'Centro-Oeste']
df_Distribution_only_region = df_Distribution_only_region[df_Distribution_only_region['State'].isin(State_to_keep)] 

race_that_want = ['White', 'Black', 'Mixed']
df_Distribution_only_region = df_Distribution_only_region[df_Distribution_only_region['Race'].isin(race_that_want)]

group = df_Distribution_only_region.groupby(['Region', 'Year'])['Quantity(10^3)'].transform('sum')


df_Distribution_only_region['Total'] = group

df_Distribution_only_region['Percentage'] = (df_Distribution_only_region['Quantity(10^3)'] / df_Distribution_only_region['Total']) * 100
df_Distribution_only_region = df_Distribution_only_region.drop(columns = ['State', 'Total'])
df_Distribution_only_region

```

```python
# Copying the first table 
table_1 = df_Distribution_only_region.copy()

# Grouping by Race and Year, and summing the quantity by race for each year
table_1 = table_1.groupby(['Race', 'Year'])['Quantity(10^3)'].agg('sum').reset_index()
table_1 = table_1.sort_values(by = 'Year', ascending = False)

# In the homicide Data Frame is lacking the 2021 year, so I am dropping the rows that have the year 2021
table_1 = table_1[table_1['Year'] != 2021]
table_1
```

```python
# Copying from the previous analysis that contains quantity of homicide cases by race over the years
table_4 = df_Homicide_only_race.copy()

# Merging two tables to calculate
homicide_rate_calc = pd.merge(table_1, table_4, on = ['Race', 'Year'])

# Homicide rate = homicide cases / population quantity
homicide_rate_calc['Homicide Rate'] = homicide_rate_calc['Quantity'] / (homicide_rate_calc['Quantity(10^3)']) * 100

# Dropping the quantity
columns_to_drop = ['Quantity(10^3)', 'Quantity']
homicide_rate_calc = homicide_rate_calc.drop(columns = columns_to_drop)
homicide_rate_calc
```
